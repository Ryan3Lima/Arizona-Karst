```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from shapely.geometry import Point
from scipy.stats import gaussian_kde
import os

# Set the working directory
path_to_sinks_dir = r"C:\Users\rl587\Documents\GitHub\Arizona-Karst\Data\Input"
os.chdir(path_to_sinks_dir)

# Read the shapefiles and CSV
sinks_poly = gpd.read_file('sinks.shp')
sinks_attributes = pd.read_csv('sinkes_filtered.csv')
study_area_bound = gpd.read_file('MRK1m_bound.shp')

# Ensure the CRS is set correctly
sinks_poly = sinks_poly.to_crs(epsg=6341)  # UTM zone 12N, NAD83 2011
study_area_bound = study_area_bound.to_crs(epsg=6341)
```

```python
# Plot the study area boundary
fig, ax = plt.subplots(figsize=(10, 10))
study_area_bound.boundary.plot(ax=ax, color='red', linewidth=2, alpha=0.5)

# Plot the sinks polygons
sinks_poly.plot(ax=ax, color='blue', edgecolor='black', alpha=0.5)

plt.title('Sinks Polygons over Study Area Boundary')
plt.xlabel('Easting (m)')
plt.ylabel('Northing (m)')
plt.grid(True)
plt.show()
```
